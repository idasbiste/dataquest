```python
import numpy as np
movie_mapper = {
    "Star Wars: Episode I  The Phantom Menace": True,
    "Star Wars: Episode II  Attack of the Clones": True,
    "Star Wars: Episode III  Revenge of the Sith": True,
    "Star Wars: Episode IV  A New Hope": True,
    "Star Wars: Episode V The Empire Strikes Back": True,
    "Star Wars: Episode VI Return of the Jedi": True,
    np.nan: False
}
counter = 1
for col in star_wars.columns[3:9]:
    star_wars[col] = star_wars[col].map(movie_mapper)
    star_wars = star_wars.rename(columns = {
           col: "seen_" + str(counter)
        });
    counter += 1
```

```python
star_wars.head(10)
```

```python
star_wars[star_wars.columns[9:15]] = star_wars[star_wars.columns[9:15]].astype(float)
```

```python
counter = 1
for col in star_wars.columns[9:15]:
    star_wars = star_wars.rename(columns = {
           col: "ranking_" + str(counter)
        });
    counter += 1
```

```python
star_wars.columns
```

```python
star_wars[star_wars.columns[9:15]].mean()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.bar(range(6), star_wars[star_wars.columns[9:15]].mean())
```

# Summary

So far, we've cleaned up the data, we've renamed some columns and computed the average ranking of each movie. We've also drawn a bar plot of that computation.

Clearly, the "old" trilogy is much more well accepted by the overall public than the most recent one.

```python
star_wars[star_wars.columns[3:9]].sum()
```

```python
plt.bar(range(6), star_wars[star_wars.columns[3:9]].sum())
```

You can always divide these results in two parts: those who have seen the original trilogy will always see the newest movies - however, since the quality of the latter is worse, it is not a surprise that the least seen movies are II and III. Definitely, the older movies are more popular than the recent ones.

```python
males = star_wars[star_wars["Gender"] == "Male"]
females = star_wars[star_wars["Gender"] == "Female"]
```

```python
plt.bar(range(6), males[males.columns[9:15]].mean())
plt.show()

plt.bar(range(6), females[females.columns[9:15]].mean())
plt.show()

plt.bar(range(6), males[males.columns[3:9]].sum(), color = 'red')
plt.show()

plt.bar(range(6), females[females.columns[3:9]].sum(), color = 'red')
plt.show()
```

Definitely, more males watched the recent movies, which might indicate they are more loyal to the movies. However, they dislike them much more than women do.
